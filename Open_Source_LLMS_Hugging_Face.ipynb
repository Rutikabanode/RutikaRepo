```python
#mport All the Required Libraries
from langchain.llms import HuggingFacePipeline
```

```python
from transformers import AutoTokenizer
```

```python
import transformers
import torch
```

```python
#model="meta-llama/Llama-2-7b-chat-hf" #https://huggingface.co/meta-llama/Llama-2-7b-chat-hf
model="daryl149/llama-2-7b-chat-hf" #if dont have access to llma2 model https://huggingface.co/daryl149/llama-2-7b-chat-hf
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

```python
prompt="What would be a good name for a company that makes iCecreams"
```

```python
print(llm.invoke(prompt))
```
