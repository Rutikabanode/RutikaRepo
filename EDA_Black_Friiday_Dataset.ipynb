##Handel Categorical Feature
Gender, Age,City

```python
#Category Gender, Category
#Convert categoriacal into numerical
pd.get_dummies(df["Gender"])
```

```python
df["Gender"]=df["Gender"].map({"F":0,"M":1})
```

```python
df.head()
```

```python
df["Age"].unique()
```

```python
pd.get_dummies(df["Age"],drop_first=True) #Not useful
```

```python
df["Age"]=df["Age"].map({"0-17":1,"18-25":2,"26-35":3,"36-45":4,"46-50":5,"51-55":6,"55+":7})
df.head()
```

```python
#we can also do lebel encoding but its not a good practice
```

```python
df_city=pd.get_dummies(df["City_Category"],drop_first=True)
df_city.head()
```

```python
df=pd.concat((df,df_city),axis=1)
df.head()
```

```python
df.drop("City_Category",axis=1,inplace=True)
```

```python
df.head()
```

```python
df["Stay_In_Current_City_Years"].unique()
```

```python
df["Stay_In_Current_City_Years"]=df["Stay_In_Current_City_Years"].str.replace("+","")
```

```python
df.head()
```

##Handeling Missing Values

```python
df.isnull().sum()
```

```python
##Focus on replacing missing values
df["Product_Category_2"].unique()
```

```python
df["Product_Category_2"].value_counts()
```

```python
###Its a descrete feature hence Replace missing values with Mode

df["Product_Category_2"].mode()[0]
```

```python
df["Product_Category_2"]=df["Product_Category_2"].fillna(df["Product_Category_2"].mode()[0])
```

```python
df["Product_Category_2"].isnull().sum() #Null values replaced
```

```python
## Product_Category_3 replace missing values
df["Product_Category_3"].value_counts()
```

```python
df.head()
```

```python
df["Product_Category_3"].isnull().sum()
```

```python
df["Product_Category_3"]=df["Product_Category_3"].fillna(df["Product_Category_3"].mode()[0])
df.head()
```
